```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ================================================================
# PATHS
# ================================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Greenhouse Climate Control\archive\Crop_recommendation.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Greenhouse Climate Control"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ================================================================
# LOAD DATA
# ================================================================
df = pd.read_csv(DATA_PATH)

label = LabelEncoder()
df['label'] = label.fit_transform(df['label'])

X = df.drop('label', axis=1)
y = df['label']

# ================================================================
# SCALING
# ================================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_scaler.pkl"))

X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


# ================================================================
# CNNâ€“LSTM MODEL BUILDER
# ================================================================
def build_model(params):
    model = Sequential([
        Conv1D(params["filters"], kernel_size=2, activation='relu',
               input_shape=(X_train.shape[1], 1)),
        MaxPooling1D(pool_size=2),
        LSTM(params["lstm_units"]),
        Dropout(params["dropout"]),
        Dense(32, activation='relu'),
        Dense(1, activation='linear')
    ])

    model.compile(optimizer=params["optimizer"], loss="mse")
    return model


# ================================================================
# EVALUATION FUNCTION
# ================================================================
def evaluate_params(params):
    model = build_model(params)

    history = model.fit(
        X_train, y_train,
        epochs=8,
        batch_size=32,
        validation_split=0.2,
        verbose=0
    )
    val_loss = history.history["val_loss"][-1]
    return val_loss


# ================================================================
#  HYBRID AIS + CSA OPTIMIZER
# ================================================================
def hybrid_ais_csa(pop_size=6, csa_iterations=5):
    population = []

    # === AIS: Generate immune antibodies ===
    for _ in range(pop_size):
        params = {
            "filters": random.choice([16, 32, 48]),
            "lstm_units": random.choice([32, 64, 96]),
            "dropout": random.choice([0.2, 0.3, 0.4]),
            "optimizer": random.choice(["adam", "rmsprop"])
        }
        population.append(params)

    # Evaluate AIS
    scores = [evaluate_params(p) for p in population]

    # Select best antibody (AIS output)
    best_idx = np.argmin(scores)
    best_params = population[best_idx]

    # === Crow Search Algorithm refinement ===
    memory = best_params.copy()

    for _ in range(csa_iterations):
        new_params = memory.copy()

        # Mutate CSA
        new_params["filters"] = random.choice([16, 32, 48])
        new_params["lstm_units"] = random.choice([32, 64, 96])
        new_params["dropout"] = random.choice([0.2, 0.3, 0.4])

        new_score = evaluate_params(new_params)
        old_score = evaluate_params(memory)

        if new_score < old_score:
            memory = new_params.copy()

    return memory


# ================================================================
# RUN HYBRID OPTIMIZATION
# ================================================================
best_params = hybrid_ais_csa()
print("\nðŸ”¥ BEST HYBRID AIS+CSA PARAMETERS:")
print(best_params)

# ================================================================
# TRAIN FINAL MODEL
# ================================================================
model = build_model(best_params)

es = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[es],
    verbose=1
)

# ================================================================
# PREDICT
# ================================================================
y_pred = model.predict(X_test).flatten()

MAE = mean_absolute_error(y_test, y_pred)
RMSE = mean_squared_error(y_test, y_pred)**0.5
R2 = r2_score(y_test, y_pred)

results = {"MAE": float(MAE), "RMSE": float(RMSE), "R2": float(R2)}

# ================================================================
# SAVE MODEL + CONFIG + RESULTS
# ================================================================
model.save(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_model.h5"))
json.dump(results, open(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_results.json"), "w"), indent=4)
yaml.dump(best_params, open(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_config.yaml"), "w"))

# ================================================================
# GRAPHS (SHOW + SAVE)
# ================================================================

# HEATMAP
plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(X).corr(), cmap="coolwarm")
plt.title("Hybrid Feature Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_heatmap.png"))
plt.show()

# LOSS GRAPH
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.title("Hybrid Model Loss Curve")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_loss_graph.png"))
plt.show()

# PREDICTION GRAPH
plt.figure(figsize=(9, 6))
plt.plot(y_test[:100].values, label="Actual")
plt.plot(y_pred[:100], label="Predicted")
plt.legend()
plt.title("Hybrid Actual vs Predicted")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_prediction_graph.png"))
plt.show()

# COMPARISON GRAPH
plt.figure(figsize=(8, 6))
plt.bar(["MAE", "RMSE", "R2"], [MAE, RMSE, R2])
plt.title("Hybrid Model Metrics")
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_comparison_graph.png"))
plt.show()

# RESULT GRAPH
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.title("Hybrid Full Prediction Result")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_smart_greenhouse_result_graph.png"))
plt.show()

print("\nâœ… HYBRID AIS + CSA Completed Successfully!")
print(results)
```
