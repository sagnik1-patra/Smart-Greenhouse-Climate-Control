```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ================================================================
# üìÇ PATHS
# ================================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Greenhouse Climate Control\archive\Crop_recommendation.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Greenhouse Climate Control"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ================================================================
# üìå LOAD DATA
# ================================================================
df = pd.read_csv(DATA_PATH)

# Encode target (crop label)
label = LabelEncoder()
df['label'] = label.fit_transform(df['label'])

X = df.drop('label', axis=1)
y = df['label']

# ================================================================
# üîß SCALE FEATURES
# ================================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Save scaler
joblib.dump(scaler, os.path.join(OUTPUT_DIR, "smart_greenhouse_scaler.pkl"))

# Reshape for CNN‚ÄìLSTM ‚Üí (samples, timesteps, features)
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# ================================================================
# üî• TRAIN/TEST SPLIT
# ================================================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ================================================================
# ü§ñ CNN + LSTM MODEL
# ================================================================
model = Sequential([
    Conv1D(32, kernel_size=2, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    LSTM(64, return_sequences=False),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse')

# Early stopping
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train
history = model.fit(
    X_train, y_train,
    epochs=60,
    batch_size=32,
    validation_split=0.2,
    callbacks=[es],
    verbose=1
)

# ================================================================
# üìà PREDICTIONS
# ================================================================
y_pred = model.predict(X_test).flatten()

# ================================================================
# üìä METRICS (FIXED)
# ================================================================
MAE = mean_absolute_error(y_test, y_pred)
RMSE = mean_squared_error(y_test, y_pred) ** 0.5  # FIXED (No squared=False)
R2 = r2_score(y_test, y_pred)

results = {
    "MAE": float(MAE),
    "RMSE": float(RMSE),
    "R2": float(R2)
}

# Save results JSON
json.dump(results, open(os.path.join(OUTPUT_DIR, "smart_greenhouse_results.json"), "w"), indent=4)

# Save config YAML
config = {
    "model": "CNN + LSTM",
    "scaler": "MinMaxScaler",
    "epochs": 60,
    "optimizer": "adam"
}
yaml.dump(config, open(os.path.join(OUTPUT_DIR, "smart_greenhouse_config.yaml"), "w"))

# Save H5 model
model.save(os.path.join(OUTPUT_DIR, "smart_greenhouse_model.h5"))

# ================================================================
# üå° HEATMAP
# ================================================================
plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(X).corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "smart_greenhouse_heatmap.png"))
plt.close()

# ================================================================
# üìâ LOSS GRAPH
# ================================================================
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Training & Validation Loss")
plt.savefig(os.path.join(OUTPUT_DIR, "smart_greenhouse_loss_graph.png"))
plt.close()

# ================================================================
# üîÆ PREDICTION GRAPH
# ================================================================
plt.figure(figsize=(9, 6))
plt.plot(y_test[:100].values, label="Actual")
plt.plot(y_pred[:100], label="Predicted")
plt.legend()
plt.title("Actual vs Predicted (First 100 Samples)")
plt.savefig(os.path.join(OUTPUT_DIR, "smart_greenhouse_prediction_graph.png"))
plt.close()

# ================================================================
# üìä COMPARISON GRAPH
# ================================================================
plt.figure(figsize=(8, 6))
metrics_names = ["MAE", "RMSE", "R2"]
metrics_values = [MAE, RMSE, R2]

plt.bar(metrics_names, metrics_values)
plt.title("Model Performance Metrics")
plt.savefig(os.path.join(OUTPUT_DIR, "smart_greenhouse_comparison_graph.png"))
plt.close()

# ================================================================
# üìà RESULT GRAPH
# ================================================================
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.title("Prediction Results (Full Test Set)")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "smart_greenhouse_result_graph.png"))
plt.close()

print("‚úÖ ALL FILES SAVED SUCCESSFULLY!")
print("üìÅ Directory:", OUTPUT_DIR)
print(json.dumps(results, indent=4))
```
