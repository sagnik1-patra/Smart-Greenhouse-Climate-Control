```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ================================================================
# ðŸ“‚ PATHS
# ================================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Greenhouse Climate Control\archive\Crop_recommendation.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Greenhouse Climate Control"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ================================================================
# ðŸ“Œ LOAD DATA
# ================================================================
df = pd.read_csv(DATA_PATH)

label = LabelEncoder()
df['label'] = label.fit_transform(df['label'])

X = df.drop('label', axis=1)
y = df['label']

# ================================================================
# ðŸ”§ SCALE FEATURES
# ================================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_scaler.pkl"))

X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# SPLIT
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ================================================================
# ðŸ¤– MODEL BUILDER (CNNâ€“LSTM)
# ================================================================
def build_model(params):
    model = Sequential([
        Conv1D(params["filters"], kernel_size=2, activation="relu",
               input_shape=(X_train.shape[1], 1)),
        MaxPooling1D(pool_size=2),
        LSTM(params["lstm_units"]),
        Dropout(params["dropout"]),
        Dense(32, activation="relu"),
        Dense(1, activation="linear")
    ])

    model.compile(optimizer=params["optimizer"], loss="mse")
    return model

# ================================================================
# ðŸ“Š PARAMETER EVALUATION
# ================================================================
def evaluate_params(params):
    model = build_model(params)
    history = model.fit(
        X_train, y_train,
        epochs=8,
        batch_size=32,
        validation_split=0.2,
        verbose=0
    )
    return history.history["val_loss"][-1]

# ================================================================
# ðŸ§¬ STAGE 1 â€” AIS GLOBAL EXPLORATION
# ================================================================
def AIS_stage(pop_size=6):
    population = []
    
    for _ in range(pop_size):
        params = {
            "filters": random.choice([16, 32, 48]),
            "lstm_units": random.choice([32, 64, 96]),
            "dropout": random.choice([0.2, 0.3, 0.4]),
            "optimizer": random.choice(["adam", "rmsprop"])
        }
        population.append(params)

    # Score antibodies
    scores = [evaluate_params(p) for p in population]

    # Select best antibody
    best_idx = np.argmin(scores)
    return population[best_idx]

# ================================================================
# ðŸ¦… STAGE 2 â€” CSA LOCAL REFINEMENT
# ================================================================
def CSA_stage(initial_params, iterations=5):
    best = initial_params.copy()

    for _ in range(iterations):
        new_params = best.copy()

        # Slight mutation
        new_params["filters"] = random.choice([16, 32, 48])
        new_params["lstm_units"] = random.choice([32, 64, 96])
        new_params["dropout"] = random.choice([0.2, 0.3, 0.4])

        old_score = evaluate_params(best)
        new_score = evaluate_params(new_params)

        if new_score < old_score:
            best = new_params.copy()

    return best

# ================================================================
# ðŸš€ RUN SEQUENTIAL HYBRIDIZATION: AIS â†’ CSA
# ================================================================
print("\nðŸ”µ Running AIS Stage...")
ais_best = AIS_stage()
print("Best AIS params:", ais_best)

print("\nðŸŸ£ Running CSA Refinement...")
best_params = CSA_stage(ais_best)
print("\nðŸ”¥ FINAL Sequential Hybrid Params:", best_params)

# ================================================================
# TRAIN FINAL MODEL
# ================================================================
model = build_model(best_params)

es = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[es],
    verbose=1
)

# ================================================================
# PREDICTIONS
# ================================================================
y_pred = model.predict(X_test).flatten()

MAE = mean_absolute_error(y_test, y_pred)
RMSE = mean_squared_error(y_test, y_pred)**0.5
R2 = r2_score(y_test, y_pred)

results = {"MAE": float(MAE), "RMSE": float(RMSE), "R2": float(R2)}

# ================================================================
# SAVE MODEL + RESULTS
# ================================================================
model.save(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_model.h5"))
json.dump(results, open(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_results.json"), "w"), indent=4)
yaml.dump(best_params, open(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_config.yaml"), "w"))

# ================================================================
# ðŸ“Š GRAPHS â€” SHOW + SAVE
# ================================================================

# HEATMAP
plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(X).corr(), cmap="coolwarm")
plt.title("Sequential Hybrid Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_heatmap.png"))
plt.show()

# LOSS CURVE
plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Sequential Hybrid Loss Curve")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_loss_graph.png"))
plt.show()

# PRED vs ACTUAL
plt.figure(figsize=(9, 6))
plt.plot(y_test[:100].values, label="Actual")
plt.plot(y_pred[:100], label="Predicted")
plt.title("Sequential Hybrid Actual vs Predicted")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_prediction_graph.png"))
plt.show()

# COMPARISON GRAPH
plt.figure(figsize=(8, 6))
plt.bar(["MAE", "RMSE", "R2"], [MAE, RMSE, R2])
plt.title("Sequential Hybrid Metrics")
plt.savefig(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_comparison_graph.png"))
plt.show()

# FULL RESULT GRAPH
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.title("Sequential Hybrid Full Prediction Results")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "sequential_smart_greenhouse_result_graph.png"))
plt.show()

print("\nâœ… SEQUENTIAL AIS â†’ CSA Hybridization Completed!")
print(results)
```
